```python

# calculate the Pearson's correlation between two variables
from numpy import mean
from numpy import std
from numpy import cov
from numpy.random import randn
from numpy.random import seed
from matplotlib import pyplot as plt
import seaborn as sns

from scipy.stats import pearsonr
# seed random number generator
seed(1)

# prepare data
data1 = 20 * randn(1000) + 100
data2 = data1 + (10 * randn(1000) + 50)

# calculate covariance matrix
covariance = cov(data1, data2)

# calculate Pearson's correlation
corr, _ = pearsonr(data1, data2)

# plot
plt.scatter(data1, data2)
plt.show()

# summarize
print('data1: mean=%.3f stdv=%.3f' % (mean(data1), std(data1)))
print('data2: mean=%.3f stdv=%.3f' % (mean(data2), std(data2)))
print('Covariance: %.3f' % covariance[0][1])
print('Pearsons correlation: %.3f' % corr)
```

```python
# Unit03 Ex1 - Covariance and Pearson Correlation with Real-World Variable Names

from numpy import mean, std, cov
from numpy.random import randn, seed
from matplotlib import pyplot as plt
from scipy.stats import pearsonr

# Seed for reproducibility
seed(42)

# Simulated temperature data (in Celsius)
temperature = 5 * randn(1000) + 25  # average around 25°C

# Ice cream sales (in £), positively correlated with temperature
ice_cream_sales = temperature * 15 + (10 * randn(1000) + 100)

# Sunglasses sales (in £), also positively correlated but more noise
sunglasses_sales = temperature * 10 + (20 * randn(1000) + 50)

# Calculate covariance and Pearson correlation between temperature and ice cream sales
cov_matrix_temp_ice = cov(temperature, ice_cream_sales)
corr_temp_ice, _ = pearsonr(temperature, ice_cream_sales)

# Calculate covariance and Pearson correlation between temperature and sunglasses sales
cov_matrix_temp_sunglasses = cov(temperature, sunglasses_sales)
corr_temp_sunglasses, _ = pearsonr(temperature, sunglasses_sales)

# Visualise the relationships
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.scatter(temperature, ice_cream_sales, alpha=0.5, color='blue')
plt.title("Temperature vs Ice Cream Sales")
plt.xlabel("Temperature (°C)")
plt.ylabel("Ice Cream Sales (£)")

plt.subplot(1, 2, 2)
plt.scatter(temperature, sunglasses_sales, alpha=0.5, color='green')
plt.title("Temperature vs Sunglasses Sales")
plt.xlabel("Temperature (°C)")
plt.ylabel("Sunglasses Sales (£)")

plt.tight_layout()
plt.show()

# Summary output
print('Temperature: mean = %.3f, std deviation = %.3f' % (mean(temperature), std(temperature)))
print('Ice Cream Sales: mean = %.3f, std deviation = %.3f' % (mean(ice_cream_sales), std(ice_cream_sales)))
print('Sunglasses Sales: mean = %.3f, std deviation = %.3f' % (mean(sunglasses_sales), std(sunglasses_sales)))

print('\n--- Temperature vs Ice Cream Sales ---')
print('Covariance: %.3f' % cov_matrix_temp_ice[0][1])
print('Pearson’s correlation: %.3f' % corr_temp_ice)

print('\n--- Temperature vs Sunglasses Sales ---')
print('Covariance: %.3f' % cov_matrix_temp_sunglasses[0][1])
print('Pearson’s correlation: %.3f' % corr_temp_sunglasses)
```
