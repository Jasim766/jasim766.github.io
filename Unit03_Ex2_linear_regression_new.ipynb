```python
import matplotlib.pyplot as plt
from scipy import stats

# Example scenario: Advertising budget vs product sales
ad_spend = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]  # in thousands (£)
product_sales = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]  # number of units sold

# Perform linear regression
slope, intercept, r, p, std_err = stats.linregress(ad_spend, product_sales)

# Pearson correlation
corr, _ = stats.pearsonr(ad_spend, product_sales)
print('Pearson’s correlation: %.3f' % corr)

# Regression prediction function
def predict_sales(ad):
    return slope * ad + intercept

# Generate model predictions
predicted_sales = list(map(predict_sales, ad_spend))

# Plot data and regression line
plt.scatter(ad_spend, product_sales, label="Actual Data")
plt.plot(ad_spend, predicted_sales, color='red', label="Regression Line")
plt.title("Advertising Spend vs Product Sales")
plt.xlabel("Advertising Spend (£1000s)")
plt.ylabel("Product Sales (Units)")
plt.legend()
plt.grid(True)
plt.show()
```
